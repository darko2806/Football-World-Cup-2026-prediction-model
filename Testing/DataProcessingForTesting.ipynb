```python
def download_fifa_ranking_snapshot(date_id):
    candidate_api_urls = [
        f"https://inside.fifa.com/api/ranking-overview?locale=en&dateId={date_id}",
        f"https://www.fifa.com/api/ranking-overview?locale=en&dateId={date_id}",
    ]

    for api_url in candidate_api_urls:
        response = session.get(api_url, timeout=30)

        content_type = response.headers.get("Content-Type", "")
        print(date_id, response.status_code, content_type, api_url)

        if response.status_code != 200:
            continue

        text_start = response.text.strip()[:30].lower()
        if text_start.startswith("<!doctype") or text_start.startswith("<html"):
            print("Skipping HTML response.")
            continue

        try:
            data = response.json()
        except Exception:
            print("Skipping invalid JSON response.")
            print(response.text[:500])
            continue

        rankings = data.get("rankings", [])

        if len(rankings) == 0:
            print("JSON response has 0 rankings.")
            print("Available keys:", data.keys())
            continue

        rows = []

        for item in rankings:
            ranking_item = item.get("rankingItem", {})

            rank = ranking_item.get("rank")
            previous_rank = ranking_item.get("previousRank")

            rows.append(
                {
                    "date_id": date_id,
                    "country": ranking_item.get("name"),
                    "country_code": ranking_item.get("countryCode"),
                    "rank": rank,
                    "previous_rank": previous_rank,
                    "rank_change": (
                        previous_rank - rank
                        if previous_rank is not None and rank is not None
                        else None
                    ),
                    "total_points": ranking_item.get("totalPoints"),
                    "previous_points": item.get("previousPoints"),
                    "confederation": item.get("tag", {}).get("text"),
                }
            )

        return pd.DataFrame(rows)

    raise RuntimeError(f"Could not download valid FIFA ranking table for date_id={date_id}")
```

```python
def download_fifa_ranking_snapshot(rank_date, date_id):
    candidate_api_urls = [
        f"https://inside.fifa.com/api/ranking-overview?locale=en&dateId={date_id}",
        f"https://www.fifa.com/api/ranking-overview?locale=en&dateId={date_id}",
    ]

    for api_url in candidate_api_urls:
        response = session.get(api_url, timeout=30)

        content_type = response.headers.get("Content-Type", "")
        print(rank_date, date_id, response.status_code, content_type, api_url)

        if response.status_code != 200:
            continue

        text_start = response.text.strip()[:30].lower()

        if text_start.startswith("<!doctype") or text_start.startswith("<html"):
            print("Skipping HTML response.")
            continue

        try:
            data = response.json()
        except Exception:
            print("Skipping invalid JSON response.")
            print(response.text[:500])
            continue

        rankings = data.get("rankings", [])

        if len(rankings) == 0:
            print("Skipping because JSON contains 0 rankings.")
            print("Available keys:", data.keys())
            continue

        rows = []

        for item in rankings:
            ranking_item = item.get("rankingItem", {})

            rank = ranking_item.get("rank")
            previous_rank = ranking_item.get("previousRank")

            rows.append(
                {
                    "rank_date": rank_date,
                    "date_id": date_id,
                    "country": ranking_item.get("name"),
                    "country_code": ranking_item.get("countryCode"),
                    "rank": rank,
                    "previous_rank": previous_rank,
                    "rank_change": (
                        previous_rank - rank
                        if previous_rank is not None and rank is not None
                        else None
                    ),
                    "total_points": ranking_item.get("totalPoints"),
                    "previous_points": item.get("previousPoints"),
                    "confederation": item.get("tag", {}).get("text"),
                    "source_url": f"https://inside.fifa.com/fifa-world-ranking/men?dateId={date_id}",
                }
            )

        df = pd.DataFrame(rows)

        df["rank_date"] = pd.to_datetime(df["rank_date"])

        df = df[
            [
                "rank_date",
                "date_id",
                "country",
                "country_code",
                "rank",
                "previous_rank",
                "rank_change",
                "total_points",
                "previous_points",
                "confederation",
                "source_url",
            ]
        ]

        return df

    raise RuntimeError(f"Could not download valid ranking for {rank_date}, {date_id}")
```

```python
import re
ranking_periods_2025_2026 = {
    "2025-04-03": "03 April 2025",
    "2025-07-10": "10 July 2025",
    "2025-09-18": "18 September 2025",
    #"2025-10-17": "17 October 2025",
    "2026-01-19": "19 January 2026",
    "2026-04-01": "01 April 2026",
}
```

```python
def parse_first_number(value):
    value = str(value)

    # Remove spaces and non-breaking spaces
    value = value.replace("\xa0", " ").strip()

    # Find numbers like:
    # 1877.6
    # 1,877.6
    # 1877
    # -12.5
    match = re.search(r"[-+]?\d+(?:,\d{3})*(?:\.\d+)?", value)

    if match is None:
        return None

    return float(match.group(0).replace(",", ""))


def parse_rank(value):
    value = str(value)
    match = re.search(r"\d+", value)

    if match is None:
        return None

    return int(match.group(0))


def parse_team_and_code(value):
    value = str(value).strip()

    # Examples this is designed to handle:
    # "Brazil (BRA)"
    # "Brazil BRA"
    # "Brazil"
    match = re.search(r"^(.*?)\s*\(([A-Z]{2,4})\)", value)

    if match is not None:
        return match.group(1).strip(), match.group(2).strip()

    match = re.search(r"^(.*?)\s+([A-Z]{2,4})$", value)

    if match is not None:
        return match.group(1).strip(), match.group(2).strip()

    return value, None
```

```python
def find_ranking_table(tables):
    for table in tables:
        cols = [str(c).lower() for c in table.columns]
        col_text = " ".join(cols)

        if "rank" in col_text and ("team" in col_text or "country" in col_text):
            return table.copy()

    raise RuntimeError("Could not find a ranking table. Print the tables to inspect them manually.")
```

```python
def clean_football_ranking_table(raw_df, rank_date, period, source_url):
    df = raw_df.copy()

    df.columns = [
        str(col)
        .replace("\xa0", " ")
        .strip()
        .lower()
        for col in df.columns
    ]

    print("Columns found:", list(df.columns))

    rank_col = [c for c in df.columns if "rank" in c][0]

    team_candidates = [c for c in df.columns if "team" in c or "country" in c]
    team_col = team_candidates[0]

    point_candidates = [c for c in df.columns if "point" in c or "pts" in c]
    current_points_col = point_candidates[0] if len(point_candidates) > 0 else None

    previous_points_col = None
    for c in point_candidates:
        if "prev" in c or "previous" in c:
            previous_points_col = c
            break

    previous_rank_col = None
    for c in df.columns:
        if ("prev" in c or "previous" in c) and "rank" in c:
            previous_rank_col = c
            break

    rows = []

    for _, row in df.iterrows():
        rank = parse_rank(row[rank_col])
        country, country_code = parse_team_and_code(row[team_col])

        if rank is None:
            continue

        total_points = (
            parse_first_number(row[current_points_col])
            if current_points_col is not None
            else None
        )

        previous_points = (
            parse_first_number(row[previous_points_col])
            if previous_points_col is not None
            else None
        )

        previous_rank = (
            parse_rank(row[previous_rank_col])
            if previous_rank_col is not None
            else None
        )

        rows.append(
            {
                "rank_date": rank_date,
                "date_id": f"football_ranking_{rank_date}",
                "country": country,
                "country_code": country_code,
                "rank": rank,
                "previous_rank": previous_rank,
                "rank_change": (
                    previous_rank - rank
                    if previous_rank is not None and rank is not None
                    else None
                ),
                "total_points": total_points,
                "previous_points": previous_points,
                "confederation": None,
                "source_url": source_url,
            }
        )

    result = pd.DataFrame(rows)

    result["rank_date"] = pd.to_datetime(result["rank_date"])

    result = result[
        [
            "rank_date",
            "date_id",
            "country",
            "country_code",
            "rank",
            "previous_rank",
            "rank_change",
            "total_points",
            "previous_points",
            "confederation",
            "source_url",
        ]
    ]

    return result
```

```python
def download_football_ranking_snapshot(rank_date, period, max_pages=5):
    period_query = period.replace(" ", "+")

    all_pages = []

    for page in range(1, max_pages + 1):
        if page == 1:
            url = f"https://football-ranking.com/ranks?period={period_query}"
        else:
            url = f"https://football-ranking.com/ranks?period={period_query}&page={page}"

        print("Reading:", url)

        tables = pd.read_html(url)
        print(f"Found {len(tables)} tables")

        raw_table = find_ranking_table(tables)

        cleaned = clean_football_ranking_table(
            raw_df=raw_table,
            rank_date=rank_date,
            period=period,
            source_url=url,
        )

        print("Page shape:", cleaned.shape)

        all_pages.append(cleaned)

        time.sleep(0.5)

    snapshot = pd.concat(all_pages, ignore_index=True)

    snapshot = (
        snapshot
        .drop_duplicates(subset=["rank"])
        .sort_values("rank")
        .reset_index(drop=True)
    )

    return snapshot
```
