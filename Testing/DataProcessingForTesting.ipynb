```python
import time
import requests
import pandas as pd
from pathlib import Path
from urllib.parse import urlparse, parse_qs
```

```python
ranking_urls = [
    "https://inside.fifa.com/fifa-world-ranking/men?dateId=id13869",
    "https://inside.fifa.com/fifa-world-ranking/men?dateId=id14233",
    "https://inside.fifa.com/fifa-world-ranking/men?dateId=id14597",
    "https://inside.fifa.com/fifa-world-ranking/men?dateId=FRS_Male_Football_20251119",
    "https://inside.fifa.com/fifa-world-ranking/men?dateId=FRS_Male_Football_20251219",
]
```

```python
def extract_date_id(url):
    parsed_url = urlparse(url)
    query = parse_qs(parsed_url.query)
    return query["dateId"][0]


date_ids = [extract_date_id(url) for url in ranking_urls]
date_ids
```

```python
session = requests.Session()

session.headers.update(
    {
        "User-Agent": (
            "Mozilla/5.0 (X11; Linux x86_64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/124.0.0.0 Safari/537.36"
        ),
        "Accept": "application/json, text/plain, */*",
        "Accept-Language": "en-US,en;q=0.9",
        "Referer": "https://inside.fifa.com/fifa-world-ranking/men",
        "Origin": "https://inside.fifa.com",
    }
)
```

```python
def download_fifa_ranking_snapshot(rank_date, date_id):
    candidate_api_urls = [
        f"https://inside.fifa.com/api/ranking-overview?locale=en&dateId={date_id}",
        f"https://www.fifa.com/api/ranking-overview?locale=en&dateId={date_id}",
    ]

    for api_url in candidate_api_urls:
        response = session.get(api_url, timeout=30)

        content_type = response.headers.get("Content-Type", "")
        print(rank_date, date_id, response.status_code, content_type, api_url)

        if response.status_code != 200:
            continue

        # Skip HTML responses
        text_start = response.text.strip()[:20].lower()
        if text_start.startswith("<!doctype") or text_start.startswith("<html"):
            print("Skipping because response is HTML, not JSON.")
            continue

        try:
            data = response.json()
        except Exception:
            print("Skipping because response is not valid JSON.")
            print(response.text[:300])
            continue

        rankings = data.get("rankings", [])

        if len(rankings) == 0:
            print("Skipping because JSON contains 0 rankings.")
            print("JSON keys:", data.keys())
            continue

        rows = []

        for item in rankings:
            ranking_item = item["rankingItem"]

            rank = ranking_item.get("rank")
            previous_rank = ranking_item.get("previousRank")

            rows.append(
                {
                    "rank_date": rank_date,
                    "date_id": date_id,
                    "country": ranking_item.get("name"),
                    "country_code": ranking_item.get("countryCode"),
                    "rank": rank,
                    "previous_rank": previous_rank,
                    "rank_change": (
                        previous_rank - rank
                        if previous_rank is not None and rank is not None
                        else None
                    ),
                    "total_points": ranking_item.get("totalPoints"),
                    "previous_points": item.get("previousPoints"),
                    "confederation": item.get("tag", {}).get("text"),
                }
            )

        return pd.DataFrame(rows)

    raise RuntimeError(f"Could not download valid JSON ranking for {rank_date}, {date_id}")
```

```python
ranking_date_ids = {
    "2022-12-22": "id13869",
    "2023-12-21": "id14233",
    "2024-12-19": "id14597",
}
```

```python
output_dir = Path("../data/fifa_rankings")
output_dir.mkdir(parents=True, exist_ok=True)

ranking_dfs = []

for rank_date, date_id in ranking_date_ids.items():
    df = download_fifa_ranking_snapshot(rank_date, date_id)

    output_path = output_dir / f"fifa_ranking_{rank_date}.csv"
    df.to_csv(output_path, index=False)

    ranking_dfs.append(df)

    print(f"Saved {output_path} with {len(df)} teams")

    time.sleep(1.0)
```
