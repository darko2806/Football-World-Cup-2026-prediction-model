```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
```

```python
data_raw = pd.read_csv('../data/results.csv')
```

```python
data_raw['tournament'].value_counts()
```

```python
world_cup_data_full = data_raw[data_raw["tournament"].str.startswith("FIFA World Cup", na=False)]
world_cup_data_full
```

```python
world_cup_data_full['date'] = pd.to_datetime(world_cup_data_full['date'])
```

```python
world_cup_data_full['year'] = world_cup_data_full['date'].dt.year
world_cup_data_full
```

```python
world_cup_data = world_cup_data_full[world_cup_data_full['year'] > 1990]
world_cup_data
```

```python
world_cup_data.to_csv('../data/ProcessedDataForTestModel/ResultsFrom1990.csv', index = False)
```
